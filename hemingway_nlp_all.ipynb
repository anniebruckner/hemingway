```python
# load nltk's SnowballStemmer as variabled 'stemmer'
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer("english")

from nltk.stem import PorterStemmer
#from nltk.tokenize import sent_tokenize, word_tokenize
ps = PorterStemmer()

all_stories = []
totalvocab_stemmed = []
totalvocab_tokenized = []

def clean_stories(file_name):
    story = open(file_name, 'r', encoding='utf-8', errors = 'ignore') # change text file each time -- need to do a better way
    story = open(file_name).read()
    story = story.lower()
    story = re.sub(r'[^\w]', ' ', story)
    story = ''.join([i for i in story if not i.isdigit()])
    story = pattern.sub('', story)
    story = "".join(l for l in story if l not in string.punctuation)
    tokens = word_tokenize(story)
    return(story)

def tokenize_and_stem(file_name):
    # first tokenize by sentence, then by word to ensure that punctuation is caught as it's own token
    #tokens = word_tokenize(file_name)
    #tokens = [word for sent in nltk.sent_tokenize(file_name) for word in nltk.word_tokenize(sent)]
    filtered_tokens = []
    # filter out any tokens not containing letters (e.g., numeric tokens, raw punctuation)
    for token in tokens:
        if re.search('[a-zA-Z]', token) and len(token) > 1: ## must be longer than 1
            filtered_tokens.append(token)
    stems = [ps.stem(t) for t in filtered_tokens]
    return stems

def tokenize_only(file_name):
    # first tokenize by sentence, then by word to ensure that punctuation is caught as it's own token
    tokens = [word.lower() for sent in nltk.sent_tokenize(file_name) for word in nltk.word_tokenize(sent)]
    filtered_tokens = []
    # filter out any tokens not containing letters (e.g., numeric tokens, raw punctuation)
    for token in tokens:
        if re.search('[a-zA-Z]', token) and len(token) > 1: ## must be longer than 1
            filtered_tokens.append(token)
    return filtered_tokens  

for i in myfiles:
    story = clean_stories(i)
    tokens = word_tokenize(story)
    frequency = nltk.FreqDist(tokens)
    
    #allwords_stemmed = tokenize_and_stem(tokens)
    #totalvocab_stemmed.extend(allwords_stemmed)  
    #allwords_tokenized = tokenize_only(story)
    #totalvocab_tokenized.extend(allwords_tokenized)
    
    all_stories.append({'Title': i, 'story': story, 'story_tokens': tokens, 'story_tokens_freq': frequency})
```
